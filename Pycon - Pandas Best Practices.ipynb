## Count the number of ratings received by each talk

```python
ted['ratings_list'] = ted.ratings.apply(lambda x: ast.literal_eval(x))
```

```python
ted.ratings_list[0]
```

```python
ted.ratings_list[0][0]['count']
```

```python
def ratings_sum(lst):
    sum_r = 0
    for i in lst:
        sum_r += int(i['count'])
    return sum_r

ted.ratings_list.apply(ratings_sum).sort_values(ascending=False)
```

```python
ted['num_ratings'] = ted.ratings_list.apply(ratings_sum)
```

```python
ted.num_ratings.describe()
```

```python
pd.DataFrame(ted.ratings_list[0])
```

```python
def sum_rat(lst):
    df = pd.DataFrame(lst)
    return df['count'].sum()

ted.ratings_list.apply(sum_rat)
```

## Which occupations deliver the funniest ted talks at an average?

```python
ted.ratings.str.contains('Funny').sum()
```

```python
ted.ratings.str.contains('Funny').value_counts()
```

```python
def funny_ratings(lst):
    for d in lst:
        if d['name'] == 'Funny':
            return d['count']

ted.ratings_list.apply(funny_ratings)
```

```python
ted['funny_rat'] = ted.ratings_list.apply(funny_ratings)
```

```python
ted['%funny_ratings'] = ted.funny_rat / ted.num_ratings
```

```python
ted['%funny_ratings'].sample(10)
```

```python
ted.sort_values(by = '%funny_ratings', ascending=False).head(20)
```

```python
ted.loc[:, ['speaker_occupation', '%funny_ratings']].sort_values(by='%funny_ratings', ascending=False).head(20)
```

## Analyze funny rate by occupation

```python
ted.groupby(by='speaker_occupation')['%funny_ratings'].agg('mean').sort_values(ascending=False).head(10)
```
