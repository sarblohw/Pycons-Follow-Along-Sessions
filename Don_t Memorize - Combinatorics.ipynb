*Example 1:*

Given digits 12345

- How many 3-digit numbers can be formed from the given digits?
    - Repetition Not Allowed
    - Repetion Allowed

Inuition - AND is multiplication, OR is addition. Here, the number of ways we can fill in the units slot is *1 or 2 or 3 or 4 or 5* hence 5 ways as per addition rule to fill in a single slot if all 5 options are available... 
Then, once we have the number of ways for each slot, we multiply them since we need slot1 and slot2 and slot3.

Answer: 

*Repetition Allowed*

_ _ _  -- here we have 5 possibilities for the first slot and 5 possibilites for the second and 5 possibilities for the third... hence 5x5x5 = 125 ways

*Repetition Not Allowed*

_ _ _  -- here we need to decrease 1 possibility every time we pick a digit since repetition is not allowed... hence 5x4x3 = 5! = 60
